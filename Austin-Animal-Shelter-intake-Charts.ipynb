### Bar Chart Showing Intake Type Counts

### Bar Chart Showing Frequency of Animal Breeds by Intake Type

```python
#copies dataset and does not override original
animal_data_copy = animal_data.copy()

#creates a new column called Count
at_counts = animal_data_copy.groupby(['Animal Type', 'Intake Type']).size().reset_index(name='Count')

at_counts = at_counts.sort_values(by='Count', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Animal Type', y='Count', hue='Intake Type', data=at_counts)
plt.title('Frequency of Animal Breeds by Intake Type')
plt.xlabel('Animal Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

### Top 30 Dog Breeds in Shelter Bar Chart

### Top 5 Dog Breeds in Shelter by Sex Upon Intake Bar Chart

```python
dog_data = animal_data[animal_data["Animal Type"] == 'Dog']

# Get the count of breed types
dog_breed_counts = dog_data['Breed'].value_counts()

# Get the top 30 cat breeds
top_5_dog_breeds = dog_breed_counts.head(5)

# Group the data by 'Breed' and 'Sex Upon Intake' and count the occurrences
dog_breeds_sex_counts = dog_data.groupby(['Breed', 'Sex upon Intake']).size().unstack()

# Select only the top 5 cat breeds from the grouped data
top_5_dog_breeds_sex_counts = dog_breeds_sex_counts.loc[top_5_dog_breeds.index]

# Create the bar chart
plt.figure(figsize=(12, 6))
top_5_dog_breeds_sex_counts.plot(kind='bar', stacked=False)
plt.title('Top 5 Dog Breeds In Shelter (by Sex Upon Intake)')
plt.xlabel('Breed')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Sex Upon Intake')
plt.show()
```

### Number of Intakes per Month from 2015-2021

```python
animal_data['DateTime'] = pd.to_datetime(animal_data['DateTime'])

# new column 'Month' for the month values extracted from the 'DateTime' column
animal_data['Month'] = animal_data['DateTime'].dt.month

# Count the number of intakes per month
intakes_per_month = animal_data.groupby('Month')['Intake Type'].count()

# Create a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=intakes_per_month.index, y=intakes_per_month.values)
plt.title('Number of Intakes per Month')
plt.xlabel('Month')
plt.ylabel('Number of Intakes')
plt.xticks(rotation=0)
plt.show()
```

### Number of Intakes Per Month by Year

```python
animal_data_copy = animal_data.copy()

# Convert the 'DateTime' column to a datetime data type
animal_data_copy['DateTime'] = pd.to_datetime(animal_data_copy['DateTime'])

# Create new columns 'Year' and 'Month' for the year and month values extracted from the 'DateTime' column
animal_data_copy['Year'] = animal_data_copy['DateTime'].dt.year
animal_data_copy['Month'] = animal_data_copy['DateTime'].dt.month

# Group by 'Year' and 'Month' and count the number of intakes
intakes_per_month = animal_data_copy.groupby(['Year', 'Month'])['Intake Type'].count().reset_index()

# Create a bar chart
plt.figure(figsize=(20, 6))
sns.barplot(x=intakes_per_month['Month'], y=intakes_per_month['Intake Type'], hue=intakes_per_month['Year'],
            palette=['#DFFF00', '#FFBF00', '#FF7F50', '#DE3163', '#9FE2BF', '#6495ED', '#CCCCFF', '#50C878', '#DA70D6'])
plt.title('Number of Intakes per Month by Year')
plt.xlabel('Month')
plt.ylabel('Number of Intakes')
plt.xticks(rotation=0)
plt.legend(title='Year', bbox_to_anchor=(1, 1))
plt.show()
```

### Number of Intakes per Month by Year

```python
animal_data_copy = animal_data.copy()

# Convert the 'DateTime' column to a datetime data type
animal_data_copy['DateTime'] = pd.to_datetime(animal_data_copy['DateTime'])

# Create new columns 'Year' and 'Month' for the year and month values extracted from the 'DateTime' column
animal_data_copy['Year'] = animal_data_copy['DateTime'].dt.year
animal_data_copy['Month'] = animal_data_copy['DateTime'].dt.month

# Group by 'Year' and 'Month' and count the number of intakes
intakes_per_month = animal_data_copy.groupby(['Year', 'Month'])['Intake Type'].count().reset_index()

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.scatterplot(x=intakes_per_month['Month'], y=intakes_per_month['Intake Type'], hue=intakes_per_month['Year'], 
               palette=['#DFFF00', '#FFBF00', '#FF7F50', '#DE3163', '#9FE2BF', '#6495ED', '#CCCCFF', '#50C878', '#DA70D6'])
plt.title('Number of Intakes per Month by Year')
plt.xlabel('Month')
plt.ylabel('Number of Intakes')
plt.xticks(rotation=0)
plt.legend(title='Year', bbox_to_anchor=(1, 1))
plt.show()
```

### Number of Intakes for 2015

```python
#converts to datetime data type
animal_data_copy['DateTime'] = pd.to_datetime(animal_data_copy['DateTime'])

#creates new columns
animal_data_copy['Year'] = animal_data_copy['DateTime'].dt.year
animal_data_copy['Month'] = animal_data_copy['DateTime'].dt.month

#grabs data from just the year indicated
intake_year = animal_data_copy[animal_data_copy['Year'] == 2015]

#grabs data from just the month indicated
intakes_month = intake_year.groupby('Month')['Intake Type'].count()

plt.figure(figsize=(12,6))
sns.lineplot(x=intakes_month.index, y=intakes_month.values, color='#9FE2BF')
plt.xlabel('Month')
plt.ylabel("Number of intakes")
plt.title('Number of intakes for 2015')
plt.xticks(rotation=0)
plt.show()
```

### Number of Intakes for 2015, 2016, 2017

```python
# Converts to datetime data type
animal_data_copy['DateTime'] = pd.to_datetime(animal_data_copy['DateTime'])

# Creates new columns
animal_data_copy['Year'] = animal_data_copy['DateTime'].dt.year
animal_data_copy['Month'] = animal_data_copy['DateTime'].dt.month

# Grabs data from year shown
intake_year_2015 = animal_data_copy[animal_data_copy['Year'] == 2015]
intakes_month_2015 = intake_year_2015.groupby('Month')['Intake Type'].count()

# Grabs data from year shown
intake_year_2016 = animal_data_copy[animal_data_copy['Year'] == 2016]
intakes_month_2016 = intake_year_2016.groupby('Month')['Intake Type'].count()

intake_year_2017 = animal_data_copy[animal_data_copy['Year'] == 2017]
intakes_month_2017 = intake_year_2017.groupby('Month')['Intake Type'].count()

# figure size
plt.figure(figsize=(12, 6))

# Plot for 2019
sns.lineplot(x=intakes_month_2015.index, y=intakes_month_2015.values, color='#40E0D0', label='2015')

# Plot for 2020
sns.lineplot(x=intakes_month_2016.index, y=intakes_month_2016.values, color='#FFBF00', label='2016')

#plot for 2021
sns.lineplot(x=intakes_month_2017.index, y=intakes_month_2017.values, color='#E34234', label='2017')

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Number of intakes')
plt.title('Number of Intakes for 2015, 2016 and 2017')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
